# **Computing Derivatives with Differentiation**

y = x^2 + 2x +2

```python
def f(mx):
  my = mx**2 + 2*mx + 2
  return my
```

```python
y = f(x)
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
_ = ax.plot(x,y)
```

## Lets find the slope where x=2, using delta method

## First lets determine whats y is

```python
f(2)
```

## Lets call that point P where (2,10)

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
plt.scatter(2,10)
_ = ax.plot(x,y)
```

## The delta mwthod uses the difference between the two points to calculate the slope. to illusrate this, lets define anotehr point $Q$  where x=5

```python
f(5)
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
plt.scatter(2,10)
plt.scatter(5,37,c='orange',zorder=3)
_ = ax.plot(x,y)
```

## To find the slope m between P and Q,

##   $$
m = \frac{change-in-y}{change-in-x}
$$

```python
m = (37-10)/(5-2)
m
```

## to plot the line that passes through Q and P, we can rearrange the equation y = mx + b, to find b

## b = y - mx

```python
b = 37 - m*5
b
```

```python
line_y = m*x + b
```

```python
10 - m *2
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
plt.scatter(2,10)
plt.scatter(5,37,zorder=3)

plt.ylim(-10,150)
plt.plot(x,line_y,c='orange')
_ = ax.plot(x,y)
```

## the closer Q comes to P, the closer slope m comes to being the true tangent of the point P. Lets demonstrate at x=2.1

```python
f(2.1)
```

```python
m = (10.61-10)/(2.1-2)
m
```

```python
b = 10.61 - m*2.1
b
```

```python
line_y = m*x + b
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
plt.scatter(2,10)
plt.scatter(2.1,10.61,zorder=3)

plt.ylim(-10,150)
plt.plot(x,line_y,c='orange')
_ = ax.plot(x,y)
```

## The closer Q becomes to P (delta x approaches 0), the clearer it becomes that slope m at that point P (2,10) = 6

#**Differentiation Demo**

```python
def diff_demo(myf,myx,delx):
  return (myf(myx+delx) - myf(myx))/delx
```

```python
deltas = [1,0.1,0.01,0.001,0.0001,0.00001,0.000001]
```

```python
for delta in deltas:
  print(diff_demo(f,2,delta))
```

```python
for delta in deltas:
  print(diff_demo(f,-1,delta))
```
