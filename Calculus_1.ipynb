$lim_x->1$ $(x^2 -1)/$$(x-1)$

```python
def my_fn(myx):
  myy = (myx**2 -1) / (myx-1)
  return myy
```

```python
my_fn(2)
```

```python
my_fn(0.9)
```

```python
my_fn(0.999)
```

```python
my_fn(1.1)
```

```python
my_fn(1.001)
```

```python
y = my_fn(x)
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')

plt.xlim(-1,5)
plt.ylim(-1,5)

plt.axvline(x=1,color='purple',linestyle='--')
plt.axhline(y=2,color='purple',linestyle='--')

_ = ax.plot(x,y)
```

lim x-> 0 in sin x / x

```python
def sinfn(mx):
  my = np.sin(mx)/ mx
  return my
```

```python
# y = sinfn(0)   will give division error
```

```python
sinfn(0.1)
```

```python
sinfn(0.001)
```

```python
sinfn(-0.1)
```

```python
sinfn(-0.001)
```

```python
y = sinfn(x)
```

```python
fig,ax = plt.subplots()

plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')

plt.xlim(-10,10)
plt.ylim(-1,2)

plt.axvline(x=0,color='purple',linestyle='--')
plt.axhline(y=1,color='purple',linestyle='--')

_ = ax.plot(x,y)
```

lim x > infinity (25/x)

```python
def inf(mx):
  my = 25 / mx
  return my
```

```python
inf(1e3)
```

```python
inf(1e6)
```

```python
y = inf(x)
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')

plt.xlim(-10,10)
plt.ylim(-300,300)

_ = ax.plot(x,y)
```

the line on y axis, is just a conneection by matplotlib but in actual we cant draw this.

```python
left_x = x[x<0]
right_x = x[x>0]
```

```python
left_y = inf(left_x)
right_y = inf(right_x)
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')

plt.xlim(-10,10)
plt.ylim(-300,300)

ax.plot(left_x,left_y,color='C0')
_ = ax.plot(right_x,right_y,color='C0')
```
