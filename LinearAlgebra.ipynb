#**Robber Sherrif problem**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,1000,40)
```

```python
d_r = 2.5 * t
```

```python
d_s = 3 * (t -5)
```

```python
# plot the graphs then

fig,ax = plt.subplots()
plt.title("a bank robber caught")
plt.xlabel("time travel")
plt.ylabel("distance traveled")
ax.set_xlim([0,40])
ax.set_ylim([0,100])

ax.plot(t,d_r,c='green')
ax.plot(t,d_s,c='brown')
plt.axvline(x=30,color='purple',linestyle='--')
_ = plt.axhline(y=75, color='purple',linestyle='--')
```

#**Solving Linear System**

solving for substitution method/


y = 3x

-5x +2y= 2  ----->>>>  y = 1 +  5x/2

```python
x = np.linspace(-10,10,1000)  #start finish n points
```

```python
y1 = 3* x
```

```python
y2 = 1 + (5*x)/2
```

```python
fig,ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim(0,3)
ax.set_ylim(0,8)
ax.plot(x,y1,color='red')
ax.plot(x,y2,color='blue')

#since i know the answer 2,6

plt.axvline(x=2,color='purple',linestyle='--')
_ = plt.axhline(6,color='purple',linestyle='--')
```

##Elimination part.

```python
y1 = -5 + (2*x)/3

y2 = (7-2*x)/5
```
